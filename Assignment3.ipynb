#### Factorial Utility( Recursive function)

#Create a reccursive function factorial(n) that:
1. Returns the factorial of n.
2. handles two edge cases: n==0and n==1
3. prints an error msg when n is negative.

Test with:
factorial(5)
factorial(0)
factorial(-3)

```python
def factorial(n):
    if n < 0:
        print("ERROR!! n can never be negative!")
        return None
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
factorial(0)
```

```python
factorial(-3)
```

### TASK 6
#### Combined Utility function:

#Write a function process_prices(prices) that:

1. Takes a list of prices.
2. Uses map() + lambda to apply 10% discount to all prices.
3. Uses filter() to keep only discounted prices that are above 300.
4. Return both lists:
      #discounted_prices
      #filtered_prices

```python
def process_prices(prices):
    discounted_prices = list(map(lambda x: x * 0.9, prices))
    filtered_prices = list(filter(lambda x: x > 300, discounted_prices))
    return discounted_prices, filtered_prices

prices = [100, 500, 900, 50, 750]

discounted_prices, filtered_prices = process_prices(prices)

print("Discounted Prices:", discounted_prices)
print("Filtered Prices:", filtered_prices)
```

#Create the following three small functions:
1. add_prices(prices_list, price) -> add price to the list
2. get_average_price(prices_list) -> return average price
3. get_max_price(prices_list) -> returns the maximum price

#Then create a simple menu:
1. Add price
2. Show avg price
3. Show highest price
4. q -> Quit

```python
def add_prices(prices_list, price):
    prices_list.append(price)
    print(prices_list)
```

```python
add_prices([100,500,900,50,750],400)
```

```python
#2.
def get_average_price(prices_list):
    s = 0
    for i in prices_list:
        s += i
        average = s/len(prices_list)
    return average
```

```python
get_average_price([100,500,900,50,750])
```

```python
#3.
def get_max_price(prices_list):
    max_price = max(prices_list)
    return max_price
```

```python
get_max_price([100,500,900,50,750])
```

```python
prices_list = [100, 500, 900, 50, 750]

while True:
    print("\nWELCOME TO THE MENU!!\nPress the function:\n1. Add Price\n2. Show Average Price\n3. Show Highest Price\n4. q for Quit")
    
    n = input("Choose the Operation: ").strip()
    
    if n == "1":
        add_prices(prices_list, 400)
        
    elif n == "2":
        get_average_price(prices_list)
        
    elif n == "3":
        get_max_price(prices_list)
        
    elif n == "4" or n == "q" or n == "Q":
        print("Thank you for using the program!!")
        break
        
    else:
        print("Invalid choice! Please select 1, 2, 3, or q.")
```
