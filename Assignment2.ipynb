```python
#Write a program that reads an integer order_amount form the user using input()
order_amount = input("Enter the amount: ")

#Ensure you convert input to a number and handle the case when the input is non-numeric by displaying an error message and exiting the program.
if order_amount.isdigit():
    order_amount = int(order_amount)
else: 
    print("Invalid input. Please enter a valid amount.")
```

```python
#Apply the following discount rules and print the final amount:
#-> order_amount >= 2000 -- 15%
#-> 1500 <= order_amount < 2000 -- 10%
#1000 <= order_amount < 1500 -- 7%
#otherwise -- 0%


if order_amount >= 2000:
    print(f"15% dicount is given on the amount: {(order_amount*15)/100}")

elif order_amount >= 1500 and order_amount < 2000:
    print(f"10% discount is given one your order: {(order_amount*10)/100}")

elif order_amount >= 1000 and order_amount < 1500:
    print(f"7% discount is given to your order: {(order_amount*7)/100}")

else:
    print("0% discount")
```

### TASK 2:
### Process Multiple Orders

#Given a list of order amounts: orders = [1200, 2500, 800, 1750, 3000], use a for loop to apply the discount rules from Task ! to each order and print a summary table showing: order_amount -> discount% -> final_amount.

```python
orders = [1200, 2500, 800, 1750, 3000]
print("Order Amount\t\tDiscount(%)\t\tFinal Amount")
print("-"*60)
for i in orders:
    # print(f"Order Amount: {i}")
    if i >= 2000:
        #15% discount
        discount = (i*15)/100
        subtotal = i - discount
        tax = (subtotal*5)/100
        print(f"{i}\t\t\t{discount}\t\t\t{subtotal+tax}")
    elif i >= 1500 and i < 2000:
        #10% discount
        discount = (i*10)/100
        subtotal = i - discount
        tax = (subtotal*5)/100
        print(f"{i}\t\t\t{discount}\t\t\t{subtotal+tax}")
    elif i >= 1000 and i < 1500:
        #7% discount
        discount = (i*7)/100
        subtotal = i - discount
        tax = (subtotal*5)/100
        print(f"{i}\t\t\t{discount}\t\t\t{subtotal+tax}")
    else:
        #0% discount
        subtotal = i
        tax = (subtotal*5)/100
        print(f"{i}\t\t\t{discount}\t\t\t{subtotal+tax}")
```

#Also compute and print the total revenue after discounts

```python
orders = [1200, 2500, 800, 1750, 3000]
print("Order Amount\t\tDiscount(%)\t\tTotal Revenue\t\tFinal Amount")
print("-"*85)
for i in orders:
    # print(f"Order Amount: {i}")
    if i >= 2000:
        #15% discount
        discount = (i*15)/100
        subtotal = i - discount
        tax = (subtotal*5)/100
        revenue = subtotal - (discount + tax)
        print(f"{i}\t\t\t{discount}\t\t\t{revenue}\t\t\t{subtotal+tax}")

    elif i >= 1500 and i < 2000:
        #10% discount
        discount = (i*10)/100
        subtotal = i - discount
        tax = (subtotal*5)/100
        revenue = subtotal - (discount + tax)
        print(f"{i}\t\t\t{discount}\t\t\t{revenue}\t\t\t{subtotal+tax}")
    elif i >= 1000 and i < 1500:
        #7% discount
        discount = (i*7)/100
        subtotal = i - discount
        tax = (subtotal*5)/100
        revenue = subtotal - (discount + tax)
        print(f"{i}\t\t\t{discount}\t\t\t{revenue}\t\t\t{subtotal+tax}")
    else:
        #0% discount
        subtotal = i
        tax = (subtotal*5)/100
        revenue = subtotal - (discount + tax)
        print(f"{i}\t\t\t{discount}\t\t\t{revenue}\t\t\t{subtotal+tax}")
```

```python
#EXTRA: Print the number of orders that received a discount(discount>0)
discount_counts = 0
for i in orders:
    if i >= 2000:
        discount_counts += 1
    elif i >= 1500 and i < 2000:
        discount_counts += 1
    elif i >= 1000 and i < 1500:
        discount_counts += 1
print(f"The number of orders that received a discount: {discount_counts}")
```

### TASK 3
#### User Menu:

```python
#Implement a simple while-loop driven menu that keeps asking the user to choose an action until they press q to quit.

#Menu options:
#1. Add order amount to a running list(use input())
#2. Show all orders and totals after applying discounts
#3. Quit
```

```python
revenue = 0
orders = []
while True:
    action = input("Choose an action:\n1. Add order amount\n2. Show all orders and totals\n3. Quit(Press q for quit)\n")
    if action == "1":
        i = input("Enter the amount: ")
        if i.isdigit():
            i = int(i)
            orders.append(i)
        else: 
            print("Invalid input. Please enter a valid amount.")
    elif action == "2":
        print("Order Amount\t\tDiscount(%)\t\tTotal Revenue\t\tFinal Amount")
        print("-"*85)
        revenue = 0
        for i in orders:
            if i >= 2000:
                discount = (i*15)/100
            elif i >= 1500 and i < 2000:
                discount = (i*10)/100
            elif i >= 1000 and i < 1500:
                discount = (i*7)/100
            else:
                discount = 0

            subtotal = i - discount
            tax = (subtotal*5)/100
            final_amount = subtotal + tax
            revenue += final_amount
            print(f"{i}\t\t\t{discount:.2f}\t\t\t{tax:.2f}\t\t\t{final_amount:.2f}")
        print("-"*60)
        print(f"Total Revenue: {revenue:.2f}")
    elif action.lower() == "q":
        print("Exiting the program. Goodbye!!")
        break
    else:
        print("Invalid option. Please choose a valid action.")
```

#Given a list of daily sales: daily = [200, 150, 0, 400, 50, -1, 300], iterate through it with a for loop and:

-> If a day's value -1, treat it as corrupted data and break the loop (stop processing)

-> If a day's value is 0, treat it as a day with no sales and continue(skip adding to revenue)

-> For valid positive sales, add to total_sales and print the running total.

```python
#Given a list of daily sales: daily = [200, 150, 0, 400, 50, -1, 300], iterate through it with a for loop and:

#-> If a day's value -1, treat it as corrupted data and break the loop (stop processing)

#-> If a day's value is 0, treat it as a day with no sales and continue(skip adding to revenue)

#-> For valid positive sales, add to total_sales and print the running total.
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0
for i in daily:
    if i == -1:
        print("Corrupted data found!! Stop processing.")
        break
    elif i == 0:
        print("No sales on this day. Skipping!!")
        continue
    else:
        total_sales += i
        print(f"Running sales: {total_sales}")
print(f"Total sales: {total_sales}")
```

```python
#Print the final total after the loop completes (or stops due to corrupted data)
print(f"The final total sales is: {total_sales}")
```
