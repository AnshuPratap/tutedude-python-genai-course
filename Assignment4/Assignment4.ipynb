### Task 1
#### Write Sales record to a file:

#Create a list of sales amounts: Sales = [1200, 450, 980, 1500, 3000]
1. Open a file named sales_data.txt in write mode
2. Write each sale on a new line in the file
3. Close the file, then reopen it and print its contents.
4. Extra: Write the data in a comma separated contents instead of separate lines.

```python
Sales = [1200, 450, 980, 1500, 3000]
#1.
f = open("sales_data.txt", "w")
#2.
for sale in Sales:
    f.write(str(sale)+"\n")
#3.
f.close()
with open("sales_data.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
#EXTRA:
with open("comma_separated_data.txt", "w") as data:
    for sale in Sales:
        content = data.write(str(sale)+",")
```

### TASK 5
#### Create Product Info File:

1. Ask the user for three product name and their prices
2. Write them into new file products.txt in this format
3. ProductName | price
4. Read the file and print each line in proper formatting

```python
prices = {"Mouse": 500, "Keyboard": 800, "Monitor": 7000, "Pendrive": 500, "Camera": 5000}

try:
    user_input = input("Enter the discount percentage: ").strip()

    if user_input == "":
        raise ValueError("Input cannot be empty.")

    discount_percentage = int(user_input)

    with open("discount_report.txt", "w") as report:
        for product, price in prices.items():
            discount = (price * discount_percentage) / 100
            report.write(f"{product} | {price} | {discount}\n")

    total_items = len(prices)
    average_discount = sum((price * discount_percentage) / 100 for price in prices.values()) / total_items

    with open("discount_report.txt", "a") as report:
        report.write(f"\nTotal Items: {total_items}\n")
        report.write(f"Average Discount Price: {average_discount:.2f}\n")

except ValueError:
    print("Please enter a valid integer for discount percentage.")
```

### TASK 7
#### Mini project - Export discounted practices

1. Create a dictionary:
prices = {"Mouse":500, "Keyboard" : 800, "Monitor" : 7000, "Pendrive" : 500, "Camera" : 5000}
2. Ask the User for a discount percentage.
3. Write discounted prices into discount_report.txt using:
        #Product | Original Price | Discounted Price
4. After writing read and print it to the terminal.
5. Extra: Write a summary at the bottom of the file:

        Total Items: X
        Average Discount Price: Y

```python
prices = {"Mouse" : 500, "Keyboard" : 800, "Monitor" : 7000, "Pendrive" : 500, "Camera" : 5000}
discount_percentage = int(input("Enter the discount percentage: "))

with open("discount_report.txt", "w") as report:
    for product, price in prices.items():
        discounted_price = (price * discount_percentage) / 100
        report.write(product + " | " + str(price) + " | " + str(discounted_price) + "\n")
with open("discount_report.txt","r") as report:
    content = report.read()
    print(content)
#Extra:
total_items = len(prices)
average_discount_price = sum((price * discount_percentage) / 100 for price in prices.values()) / total_items
with open("discount_report.txt", "a") as report:
    report.write(f"\nTotal items: {total_items}" + "\n" + f"Average discount price: {average_discount_price:.2f}\n")
```
