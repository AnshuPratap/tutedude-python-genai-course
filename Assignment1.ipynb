### TASK 2:
#### Categories(Sets)

```python
#From your products list, create a set of categories called categories_set.(If product names do not contains categories, create a short parallel list categories = [...] with matching length and use that.)
categories = ["Hair product", "Body product", "Cooking product", "Wooden Appliance", "Mobile Appliance", "Digital Appliance", "Fruit product", "Vegetable product"]
categories_set = set(categories)
print(categories_set)
```

```python
#Demonstrate adding a new category to the set and show that duplicates are ignored.
categories_set_length = len(categories_set)
print(categories_set_length)
categories_set.add("Body product")
categories_set.add("Cooking product")
update_categories_set = categories_set
print(update_categories_set)
update_categories_set_length = len(update_categories_set)
print(update_categories_set_length)

#Checking after adding categories("Body product" and "Cooking product") the length of update_categories_set is same as of categories_set using double equals to assignment operator.

print(categories_set_length == update_categories_set_length)

#if True it proves that duplicate elements are ignored
```

```python
#Show how to check whether a category exists in a set(print a boolean result)!!
print("Body product" in categories_set)
print("Chemical products" in categories_set)
#Using membership operator ("in") we can check whether any element present or not in a set.
```

```python
#Extra: show how to get a total number of unique categories using a set.
#Using len() method we can find the number of unique elements in a set!!

print(f"The total number of categories in a set is : {len(categories_set)}")
```

#1. Using the products list and price_dict, create a list of tuples named catalog where each tuple is(product_name, price, category).

```python
update_dict = {"Shampoo":{"price":10,"category":"Personal Care"},
               "Hair Oil":{"price":150,"category":"Personal Care"},
               "Furniture":{"price":1000,"category":"Home"},
               "Chair":{"price":800,"category":"Home"},
               "Phone":{"price":40000,"category":"Electronics"},
               "Laptop":{"price":90000,"category":"Electronics"},
               "Headphones":{"price":2500,"category":"Electronics"},
               "Jeans":{"price":700,"category":"Clothing"},
               "Jacket":{"price":2000,"category":"Clothing"},
               "Book":{"price":350,"category":"Education"}}
```

```python
update_dict
```

```python
for product,details in update_dict.items():
    for key,value in details.items():
        catalog = (product,key,value)
        print(catalog)
```

#2.From catalog, create a new dictionary category_to_products that maps each category to a list of product names in that category

```python
category = [i["category"] for i in update_dict.values()]
# print(set(category))
category_to_products = {cat:[product for product,details in update_dict.items() if details["category"] == cat] for cat in set(category)}
print(category_to_products)
```

#3. Print all products that belong to the category hat has maximum number of products.

```python
category_counts = {cat:len(prod) for cat,prod in category_to_products.items()}
print(category_counts)
max_products = [prod for cat,prod in category_to_products.items() if len(prod) == max(category_counts.values())]
print(f"Category {max(category_counts, key=category_counts.get)} has the maximum products: {max_products}")
```
